```python
from matminer.datasets import get_available_datasets #This shows the available datasets that can be accessed from matminer's repository

get_available_datasets()
```

```python
from matminer.datasets import load_dataset #This downloads the specific dataset requested from matminer's online repository

et = load_dataset("elastic_tensor_2015")

et.head() #This shows the first 5 contents of the dataset
```

```python
from matminer.featurizers.conversions import StrToComposition

stc = StrToComposition()

et = stc.featurize_dataframe(et, "formula", pbar=False) 
 
# The formula column above contain simple strings,
#now we'd convert to pymatgen compositions uisng a 'conversion featurizer' which can then be used with any of the other featurizers

stc.feature_labels

et.head () #Now we see a new composition column where we can use for additional featurizers, eg, getting magpie data
```

```python
from matminer.featurizers.structure import DensityFeatures #Now, I am further featurizing using the structure column, to get density, vpa, and packing fraction

densityf = DensityFeatures()

etdf = densityf.featurize_dataframe(et, "structure")

densityf.feature_labels() #This shows us what we will be getting, that is what the featurizer can do
```

```python
from matminer.featurizers.composition import ElementProperty #Once again, we will featurize further to get the magpie data from the composition/ElementProperty module

etMag = ElementProperty.from_preset('magpie')

etdfMag = etMag.featurize_dataframe (etdf, "composition")

etdfMag.head()
```

```python
#To make a model for predicting bulk modulus, 
# Firstly, we need to have our target values in a format scikit learn can understand
#You can parse in panda seires, i.e, just the columns if you want,
# But in this case, we will be getting the numpy array out of it using the ".values" attribute object

y = etdfMag['K_VRH'].values # make a called variable called "y", set it to our etf dataframe, index it according to our target values "K_VRH"

print(y) # This prints our target values
```

```python
#Now, we need to get our training features as we are selecting a matrix and not a single vector


X = etdfMag.drop(["material_id", "nsites", "space_group", "volume", "elastic_anisotropy", "G_Reuss", "G_VRH", "G_Voigt", "K_Reuss", "K_Voigt", "poisson_ratio", "compliance_tensor", "elastic_tensor", "elastic_tensor_original", "cif", "kpoint_density", "poscar", "structure", "formula", "composition", "K_VRH"], axis=1) 

# This makes a new variable 'X' which is our training features, 
# and drops unnecessary columns (notice, we dropped the columns previously used to generate features/descriptors above) which include our target value K_VRH, 
# as we are trying to map the learning features we generated onto our target values "K_VRH" 
# We specify axis=1 to make sure we are dropping the columns


print("There are {} possible descriptors:".format(len(X.columns))) # This gives the length of x columns left in the feature matrix

print(X.columns) # This will show the number of possible descriptors or features(represented by a column each)
```

```python
# Now we will train our model using scikit learn and the features we have made ready

# And we will be training a RandomForestRegressor model here

# We will first train on the data to evaluate the training error, then run a cross validation where we estimate the error more accurately
```

```python
from sklearn.ensemble import RandomForestRegressor 

rf = RandomForestRegressor(n_estimators=100, random_state=1) 

# The n estimator parameter above is an example of a hyper parameter in a ML model, 
# so you set it before fitting your model
# You might have to try a bunch of parameters before getting the optimal estimation

# the second argument, random state = 1 is to make sure we all get the same answers 

RandomForestRegressor?
```

```python
# from sklearn.neural_network import MLPClassifier .. we can also use NN's for training our ML models instead of random forest regression above
# although NN is not appropriate for this, as our dataset is small
```

```python
# from sklearn.kernel_ridge import KernelRidge .. or kernel ridge regression, KRR for training our ML models instead of random forest regression above
```

```python
#from sklearn.gaussian_process import GaussianProcessRegressor .. uses GPR for training our ML model
```

```python
#from sklearn.linear_model import LinearRegression .. uses the basic LR for training our model
```

```python
rf.fit(X, y) 

# All scikit learn models have a '.fit' method which enables us to predict the target propoerty
```

```python
# Now lets see what our predictions look like, evaluate our model's performance

# Similar to fit, all scikit learn models have a predict method

y_pred = rf.predict(X)
#now we have a vector of our predicted values, this is where the actual machine learning takes place
```

```python
# now lets examine our training errors using a few things from scikit learn and numpy


import numpy as np
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y, y_pred)
print('rf training RMSE = {:.3f} GPa'.format(np.sqrt(mse))) #This is just a formatting syntax
```

```python
# Our training error is as shown above, which is not really our actual error as our goal with this model is to apply it 

#to new samples which has not yet been introduced.

#All we have done here is to train these samples and so this training RMSE way underestimates what our training error will be
```

```python
# Cross validation will much more accurately evaluate our errors

# Here, we run a number of independent experiments called folds, where we separate our training and testing data for assessing error

# So, in one fold, we might take 90% of a data for training, and keep 10% for testing

#We then repeat it K times e.g 10 times , and then we get 10 scores showing how each experiment did

# 10 is a pretty standard number for cross validation
```

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=10, random_state=1, shuffle=True) 

#n=10 is a standard number for cross validation, the shuffle argument will randomly shuffle the data and make sure we get same answers
```

```python
#this will give our scores on each experiments, but what we really need is the average or mean of each of these errors 

from sklearn.model_selection import cross_val_score

scores = cross_val_score(rf, X, y, scoring='neg_mean_squared_error', cv=kfold)

rmse_scores = [np.sqrt(abs(s)) for s in scores] #this is a new variable called rmse_scores with a list comprehension
print('rf Mean RMSE: {:.3f}'.format(np.mean(rmse_scores)))
```

```python
# As seen above, this almost triples the previous training error, but it's still not so bad in itself
```

```python
# Another way to evaluate our errors is by looking at some grpahs made with plotly

# First, we put some predictions on our dataframe for every fold, we use cross val function for this

from sklearn.model_selection import cross_val_predict

y_pred = cross_val_predict(rf, X, y, cv=kfold) # we don't need the scoring parameter here as it only returns the raw parameters and not the scores as done above
```

```python
etdfMag["K_VRH predicted"] = y_pred
etdfMag["percentage_error"] = (etdfMag["K_VRH"] - etdfMag["K_VRH predicted"]).abs()/etdfMag["K_VRH"] * 100

etdfMag
```

```python
# To look at it visually, we use plotly. So I recommend you read through its documentation.
```

```python
import plotly.express as px
import plotly.graph_objects as go

reference_line = go.Scatter(
    x=[0, 400],
    y=[0, 400],
    line=dict(color="black", dash="dash"),
    mode="lines",
    showlegend=False
)

fig = px.scatter(
    etdfMag, 
    x="K_VRH", 
    y="K_VRH predicted", 
    hover_name="formula", 
    color="percentage_error", 
    color_continuous_scale=px.colors.sequential.Bluered,
)

fig.add_trace(reference_line)
fig.show()
```

```python
# From the plot above, we can see that there are some outliers, e.g, carbon and water (colored red), we can tune the model hyper parameters and the training features to improve the model's prediction
```

```python
rf.feature_importances_ #this is the same as LASSO in ramprasad's video, to see the most important features that influenced the regression model. It is a numpy array, so we'd convert to plotly
```

```python
importances = rf.feature_importances_
included = X.columns.values
indices = np.argsort(importances)[::-1]

fig_bar = px.bar(
    x=included[indices][0:5], 
    y=importances[indices][0:5], 
    title="Feature Importances of Random Forest",
    labels={"x": "Feature", "y": "Importance"}
)
fig_bar.show()

#the results kind of intuitively makes sense as we are predicting the bulk modulus
```

Now, lets use Kernel Ridge Regression to train the elastic tensor dataset I just featurized

```python
from sklearn.kernel_ridge import KernelRidge

KernelRidge?
```

```python
krr = KernelRidge(alpha=1)
```

```python
krr.fit(X,y)
```

```python
y_pred2 = krr.predict(X)
```

```python
# now lets examine our training errors using a few things from scikit learn and numpy


import numpy as np
from sklearn.metrics import mean_squared_error

mse2 = mean_squared_error(y, y_pred2)
print('krr training RMSE = {:.3f} GPa'.format(np.sqrt(mse2))) #This is just a formatting syntax
```

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=10, random_state=1, shuffle=True) 

#n=10 is a standard number for cross validation, the shuffle argument will randomly shuffle the data and make sure we get same answers
```

```python
#this will give our scores on each experiments, but what we really need is the average or mean of each of these errors 

from sklearn.model_selection import cross_val_score

scores = cross_val_score(krr, X, y, scoring='neg_mean_squared_error', cv=kfold)

rmse_scores2 = [np.sqrt(abs(s)) for s in scores] #this is a new variable called rmse_scores with a list comprehension
print('Krr Mean RMSE: {:.3f}'.format(np.mean(rmse_scores2)))
```

```python
# Another way to evaluate our errors is by looking at some grpahs made with plotly

# First, we put some predictions on our dataframe for every fold, we use cross val function for this

from sklearn.model_selection import cross_val_predict

y_pred2 = cross_val_predict(krr, X, y, cv=kfold) # we don't need the scoring parameter here as it only returns the raw parameters and not the scores as done above
```

```python
etdfMag["K_VRH predicted"] = y_pred2
etdfMag["percentage_error"] = (etdfMag["K_VRH"] - etdfMag["K_VRH predicted"]).abs()/etdfMag["K_VRH"] * 100

etdfMag
```

```python
import plotly.express as px
import plotly.graph_objects as go

reference_line = go.Scatter(
    x=[0, 400],
    y=[0, 400],
    line=dict(color="black", dash="dash"),
    mode="lines",
    showlegend=False
)

fig = px.scatter(
    etdfMag, 
    x="K_VRH", 
    y="K_VRH predicted", 
    hover_name="formula", 
    color="percentage_error", 
    color_continuous_scale=px.colors.sequential.Bluered,
)

fig.add_trace(reference_line)
fig.show()
```

```python
#This Kernel Ridge Regression also does well to learn, but not as good as the RF as it's RMSE is higher.
#We have more outliers here in krr than there was in rf, (water, Na, CsBr, Carbon, Bi, etc)
#Kernel Ridge does not have feature importances as it is not based on the decision tree model which rf works with.
```
